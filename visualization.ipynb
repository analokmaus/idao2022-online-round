```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
exp_path = Path('results/exp_12')
```

```python
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
for i in range(5):
    log_fold = pd.read_csv(exp_path/f'fold_{i}.csv')
    log_fold.plot(x='epoch', y='trn_loss', kind='line', ax=ax1, grid=True, label=f'fold{i+1}')
    log_fold.plot(x='epoch', y='trn_energy_within_threshold', kind='line', ax=ax2, grid=True, label=f'fold{i+1}')
```
